```python
import json
import pandas as pd

def select_team_xi(team_id, venue_id, players_df, venues_df, xi_size=11):
    venue = venues_df.loc[venues_df['venue_id'] == venue_id].iloc[0]
    spin_fav = venue['spin_favorability']
    pace_fav = venue['pace_favorability']

    team_players = players_df[players_df['team_id'] == team_id]

    batsmen = team_players[team_players['role'].str.contains('Batsman', case=False)]
    wicketkeepers = team_players[team_players['role'].str.contains('Wicketkeeper', case=False)]
    spinners = team_players[team_players['bowling_type'].str.contains('spin', case=False)]
    pacers = team_players[team_players['bowling_type'].str.contains('fast|medium', case=False) & ~team_players['bowling_type'].str.contains('spin', case=False)]

    batsmen = batsmen.sort_values(by='recent_batting_form', ascending=False)
    spinners = spinners.sort_values(by='recent_bowling_form', ascending=False)
    pacers = pacers.sort_values(by='recent_bowling_form', ascending=False)
    wicketkeepers = wicketkeepers.sort_values(by='recent_batting_form', ascending=False)

    selected = []
    if not wicketkeepers.empty:
        wk_player_id = wicketkeepers.iloc[0].player_id
        selected.append(wk_player_id)
    else:
        wk_player_id = None

# Exclude the wicketkeeper from batsmen selection to avoid duplication
    if wk_player_id is not None:
        batsmen = batsmen[~(batsmen['player_id'] == wk_player_id)]

    batsmen_needed = max(4, xi_size - len(selected) - 5)  # adjusting for bowlers
    selected += batsmen.head(batsmen_needed)['player_id'].tolist()


    # Decide spinner and pacer count proportional to venue
    total_bowlers = xi_size - len(selected)
    spinner_slots = int(total_bowlers * (spin_fav / (spin_fav + pace_fav)))
    pacer_slots = total_bowlers - spinner_slots

    selected += spinners.head(spinner_slots)['player_id'].tolist()
    selected += pacers.head(pacer_slots)['player_id'].tolist()

    # Fill any remaining slots
    if len(selected) < xi_size:
        remaining_slots = xi_size - len(selected)
        remaining_players = team_players[~team_players['player_id'].isin(selected)].copy()
        remaining_players['combined_form'] = remaining_players['recent_batting_form'] + remaining_players['recent_bowling_form']
        remaining_players = remaining_players.sort_values(by='combined_form', ascending=False)
        selected += remaining_players.head(remaining_slots)['player_id'].tolist()

    # Convert IDs to native int for JSON serialization
    selected_players_json = json.dumps([int(pid) for pid in selected])

    spinner_count = len([pid for pid in selected if pid in spinners['player_id'].values])
    pacer_count = len([pid for pid in selected if pid in pacers['player_id'].values])

    selected_batsmen = batsmen[batsmen['player_id'].isin(selected)]
    left_right_combo = (selected_batsmen['batting_hand'].str.lower() == 'left').any() and (selected_batsmen['batting_hand'].str.lower() == 'right').any()

    return {
        'team_id': team_id,
        'venue_id': venue_id,
        'selected_players': selected_players_json,
        'spinner_count': spinner_count,
        'pacer_count': pacer_count,
        'left_right_combo': left_right_combo
    }

# Example: Generate team XIs for all 8 teams and all 4 venues
def generate_all_team_xis(players_df, venues_df, team_ids, venue_ids):
    all_selections = []
    team_xi_id = 1
    for venue_id in venue_ids:
        for team_id in team_ids:
            selection = select_team_xi(team_id, venue_id, players_df, venues_df)
            selection['team_xi_id'] = team_xi_id
            team_xi_id += 1
            all_selections.append(selection)
    return all_selections

# Assuming you have players_df including all 8 teams and venues_df including 4 venues
team_ids = [1,2,3,4,5,6,7,8]       # Your 8 teams
venue_ids = venues_df['venue_id'].tolist()  # All 4 venues

all_team_xis = generate_all_team_xis(players_df, venues_df, team_ids, venue_ids)

# Convert list of dicts to DataFrame
team_xi_df = pd.DataFrame(all_team_xis)

print(team_xi_df)
```

```python
import random

def simulate_match(team1_id, team2_id, teams_df):
    team1_strength = teams_df.loc[teams_df['team_id'] == team1_id, 'overall_strength'].values[0]
    team2_strength = teams_df.loc[teams_df['team_id'] == team2_id, 'overall_strength'].values[0]

    base_score = 150
    # Simulate first innings score with noise weighted by team strength
    team1_score = int(random.gauss(base_score * team1_strength, 15))
    team2_score = int(random.gauss(base_score * team2_strength, 15))

    # Ensure scores are positive
    team1_score = max(50, team1_score)
    team2_score = max(50, team2_score)

    if team1_score > team2_score:
        winner_id = team1_id
        margin = f"{team1_score - team2_score} runs"
    else:
        winner_id = team2_id
        margin = f"{10 - random.randint(1, 9)} wickets"  # Simplified wicket margin

    win_prob = team1_strength / (team1_strength + team2_strength)

    return {
        'team1_score': team1_score,
        'team2_score': team2_score,
        'winner_id': winner_id,
        'margin': margin,
        'win_probability': win_prob
    }
```
